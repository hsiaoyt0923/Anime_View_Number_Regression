```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import pandas as pd
```

```python
df = pd.read_csv('./datasource/BaHaMut_9.csv')
p = df['平均觀看數(萬)'].to_list()
q = []
for number in p:
    if number >= 40:
        q.append(3)
    elif 40 > number >= 20:
        q.append(2)
    elif 20 > number >= 10:
        q.append(1)
    elif number < 10:
        q.append(0)

df['人氣度'] = q
df1 = df[['作品分類1', '作品分類2', '作品分類3', '作品分類4',
          '作品分類5', '作品分類6', '原作載體', '新續作', '人氣度']]
df1
```

```python
test_list = []
for i in range(438):
    tagStr = []
    tags = df1.iloc[i].tolist()
    for tag in tags:
        if type(tag) == str:
            tagStr.append(tag)
    tagStr = ' '.join(tagStr)
    test_list.append(tagStr)
test_list
```

```python
# cv = CountVectorizer()
tf = TfidfVectorizer()
data = tf.fit_transform(test_list)
array = data.toarray()
print(array.shape)
print(array)
print(tf.get_feature_names_out())
```

```python
for _ in range(10):
    x_train, x_test, y_train, y_test = train_test_split(array, df1.人氣度, test_size=0.2)
    dec = DecisionTreeClassifier(max_depth=3)
    dec.fit(x_train, y_train)
    score = dec.score(x_test, y_test)
    print(score)
```
