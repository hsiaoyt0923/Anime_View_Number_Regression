```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./BaHaMut_7.csv')
df['人氣指標'] = round(df['平均觀看數'], ndigits=-4) // 10000
df[['導演監督', '製作廠商', '作品分類1', '作品分類2', '作品分類3',
    '作品分類4', '作品分類5', '作品分類6', '原作載體', '新續作', '人氣指標']]
df
```

```python
p = df.人氣指標.to_list()
q = []
for number in p:
    if number >= 40:
        q.append(4)
    elif 40 >= number > 20:
        q.append(3)
    elif 20 >= number > 10:
        q.append(2)
    elif 10 >= number > 1:
        q.append(1)
    elif number <= 1:
        q.append(0)
```

```python
df['人氣度'] = q
```

```python
df1 = df[['導演監督', '製作廠商', '作品分類1', '作品分類2', '作品分類3',
          '作品分類4', '作品分類5', '作品分類6', '原作載體', '新續作', '人氣度']]
df1.to_csv('dataset_for_NaiveBayes.csv', index=False)
```
